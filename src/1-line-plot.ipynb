### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

Made a line plot with x=0 to 10 and y=square of x

Added title , labels for both the axes and a grid to the plot

```python
def create_line_plot(x, y):
    try:
        if x is None or y is None or not x or not y:
            raise ValueError("Inputs x and y cannot be None or empty.")
        if len(x) != len(y):
            raise ValueError("x and y must have the same length.")
        if not all(isinstance(i, (int, float)) for i in x + y):
            raise TypeError("x and y must contain only numeric values.")
        
        plt.figure(figsize=(10, 6))
        plt.plot(x, y, marker='o', color='b', linestyle='-')
        plt.title('Line Plot: y = x²')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.grid(True)
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
create_line_plot(x, y)
```
