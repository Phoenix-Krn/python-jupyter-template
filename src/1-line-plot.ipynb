### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

Made a line plot with x=0 to 10 and y=square of x

Added title , labels for both the axes and a grid to the plot

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,11,1)
y=x**2

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='b', marker='o', linestyle='-')

plt.title('Line Plot of x=0 to 10 and y=square of x', fontsize=14)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values (Square of X)', fontsize=12)

plt.grid(True)

plt.show()
```
