### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
def create_subplots():
    try:
        x = [0, 1, 2, 3, 4]
        y = [0, 1, 4, 9, 16]
        categories = ['A', 'B', 'C', 'D']
        values = [5, 3, 9, 6]
        labels = ['Apple', 'Banana', 'Cherry', 'Date']
        sizes = [10, 15, 7, 5]

        if x is None or y is None or not x or not y:
            raise ValueError("Inputs x and y cannot be None or empty.")
        if not all(isinstance(i, (int, float)) for i in x + y):
            raise TypeError("X and Y must contain only numeric values.")
        
        fig, axs = plt.subplots(2, 2, figsize=(12, 10))

        # Line Plot
        axs[0, 0].plot(x, y)
        axs[0, 0].set_title('Line Plot')

        # Bar Plot
        axs[0, 1].bar(categories, values)
        axs[0, 1].set_title('Bar Plot')

        # Scatter Plot
        axs[1, 0].scatter(x, y)
        axs[1, 0].set_title('Scatter Plot')

        # Pie Chart
        axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
        axs[1, 1].set_title('Pie Chart')

        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

create_subplots()
```

Successfully created 2X2 subplot.
