### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['c', 'black', 'pink', 'orange']

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=colors)

plt.title('Bar Plot of Categories and Values', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)

for i, value in enumerate(values):
    plt.text(i, value + 0.2, str(value), ha='center', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Successfully created a bar plot and added colors for each bar
