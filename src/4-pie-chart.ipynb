### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
def create_pie_chart(labels, sizes):
    try:
        if labels is None or sizes is None or not labels or not sizes:
            raise ValueError("Labels and sizes cannot be None or empty.")
        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes must have the same length.")
        if not all(isinstance(size, (int, float)) and size > 0 for size in sizes):
            raise ValueError("Sizes must be positive numeric values.")
        
        plt.figure(figsize=(8, 8))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
        plt.title('Pie Chart of Programming Languages')
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
create_pie_chart(labels, sizes)
```

Successfully created a pie plot and displayed the percentage using autopct
