### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

labels = ['food', 'water', 'clothes', 'money']
sizes = [10, 30, 20, 30]

plt.figure(figsize=(10, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen', 'orange'])

plt.title('Basic Human needs', fontsize=24)

plt.show()
```

Successfully created a pie plot and displayed the percentage using autopct
